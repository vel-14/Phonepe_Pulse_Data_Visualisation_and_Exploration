```python
import os
import json
import pandas as pd
import mysql.connector
import plotly.express as px
import requests
```

```python
#aggregated_insurance_files

path1="/Users/velmurugan/Desktop/velu/python_works/phonepe_project/pulse/data/aggregated/insurance/country/india/state/"
agg_ins_states=os.listdir(path1)

#states=[state for state in cur_states if state!=".DS_Store"]

df_data1={"states":[],"years":[],"quarters":[],"transaction_type":[],"transaction_count":[],"transaction_amount":[]}

for state in agg_ins_states:
    acquired_state1=path1+state+"/"
    agg_ins_years=os.listdir(acquired_state1)

    for year in agg_ins_years:
        acquired_year1=acquired_state1+year+"/"
        agg_ins_files=os.listdir(acquired_year1)

        for file in agg_ins_files:
            acquired_file1=acquired_year1+file
            data1=open(acquired_file1,"r")
            file1=json.load(data1)
            
            for i in file1["data"]["transactionData"]:
                name=i["name"]
                count=i["paymentInstruments"][0]["count"]
                amount=i["paymentInstruments"][0]["amount"]
                df_data1["states"].append(state)
                df_data1["years"].append(year)
                df_data1["quarters"].append(int(file.strip(".json")))
                df_data1["transaction_type"].append(name)
                df_data1['transaction_count'].append(count)
                df_data1["transaction_amount"].append(amount)

aggregated_insurance = pd.DataFrame(df_data1)

aggregated_insurance['states']=aggregated_insurance["states"].replace("andaman-&-nicobar-islands","Andaman & Nicobar")
aggregated_insurance['states']=aggregated_insurance["states"].str.replace("-"," ")
aggregated_insurance['states']=aggregated_insurance["states"].str.title()
aggregated_insurance['states']=aggregated_insurance["states"].replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
```

```python
#aggregated_user_files

path3="/Users/velmurugan/Desktop/velu/python_works/phonepe_project/pulse/data/aggregated/user/country/india/state/"
agg_user_states=os.listdir(path3)

df_data3={"states":[],"years":[],"quarters":[],"brands":[],"transaction_count":[],"percentage":[]}

for state in agg_user_states:
    acquired_state3=path3+state+"/"
    agg_user_years=os.listdir(acquired_state3)

    for year in agg_user_years:
        acquired_year3=acquired_state3+year+"/"
        agg_user_files=os.listdir(acquired_year3)

        for file in agg_user_files:
            acquired_file3=acquired_year3+file
            data3=open(acquired_file3,"r")
            file3=json.load(data3)

            try:
                for i in file3['data']["usersByDevice"]:
                    brand=i["brand"]
                    count=i['count']
                    percentage=i['percentage']
                    df_data3["states"].append(state)
                    df_data3["years"].append(year)
                    df_data3["quarters"].append(int(file.strip(".json")))
                    df_data3["brands"].append(brand)
                    df_data3['transaction_count'].append(count)
                    df_data3["percentage"].append(percentage)
            except:
                pass

aggregated_user=pd.DataFrame(df_data3)

aggregated_user['states']=aggregated_user["states"].replace("andaman-&-nicobar-islands","Andaman & Nicobar")
aggregated_user['states']=aggregated_user["states"].str.replace("-"," ")
aggregated_user['states']=aggregated_user["states"].str.title()
aggregated_user['states']=aggregated_user["states"].replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
```
